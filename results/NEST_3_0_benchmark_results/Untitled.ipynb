```python
pattern = {'Network Building_Memory_thisjob_before: ': 'ini_build_mem',
           'Network Building_ru_maxrss-before: ': 'ini_build_mem_ru_maxrss',
           'Network Building_Memory_thisjob_after: ': 'net_build_mem',
           'Network Building_ru_maxrss-after: ': 'net_build_mem_ru_maxrss',
           'Network Building_Elapse_Time:': 'net_build_time',
           'Simulation_Memory_thisjob_before (10 ms): ': 'before_pre_sim_mem',
           'Simulation_ru_maxrss-before (10 ms): ': 'before_pre_sim_mem_ru_maxrss',
           'Simulation_Memory_thisjob_after (10 ms): ': 'after_pre_sim_mem',
           'Simulation_ru_maxrss-after (10 ms): ': 'after_pre_sim_mem_ru_maxrss',
           'Simulation_Elapse_Time (10 ms): ': 'pre_sim_time',
           'Simulation_Memory_thisjob_before (3000 ms): ': 'before_sim_mem',
           'Simulation_ru_maxrss-before (3000 ms): ': 'before_sim_mem_ru_maxrss',
           'Simulation_Memory_thisjob_after (3000 ms): ': 'sim_mem',
           'Simulation_ru_maxrss-after (3000 ms): ': 'sim_mem_ru_maxrss',
           'Simulation_Elapse_Time (3000 ms): ': 'sim_time'}
kernel_pattern = {'local_num_threads': 'local_num_threads',
                  'local_spike_counter': 'local_spike_counter',
                  'buffer_size_spike_data': 'buffer_size_spike',
                  'buffer_size_target_data': 'buffer_size_target',
                  'max_buffer_size_spike_data': 'max_buffer_size_spike',
                  'max_buffer_size_target_data': 'max_buffer_size_target',
                  'network_size': 'network_size',
                  'num_connections': 'num_connections',
                  'num_processes': 'num_processes',
                  'sort_connections_by_source': 'sort_connections_by_source',
                  'total_num_virtual_procs': 'total_num_virtual_procs'}
data = {}

for new_name in pattern.values():
    data[new_name] = []
for new_name in kernel_pattern.values():
    data[new_name] = []

with open('../../../../../Downloads/performance.txt') as infile:
    for line in infile:
        for pttrn in pattern.keys():
            if pttrn in line:
                data[pattern[pttrn]].append(line[len(pttrn):-1])  # do not add newline \n
        if line.startswith('{'):
            d = ast.literal_eval(line)
            for krnl_pttrn in kernel_pattern.keys():
                data[kernel_pattern[krnl_pttrn]].append(d[krnl_pttrn])

df = pd.DataFrame(data=data)
df
```
